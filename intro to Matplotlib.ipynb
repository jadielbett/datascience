## Matplotlib Workflow

```python
# 0. import matplotlib amd get it ready in jupyter notebook
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare Data
x = [1,2,3,4]
y = [20,40,60,80]

# 2. setup plot
fig, ax = plt.subplots(figsize = (10,10))

# 3. plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="Simple Plot",
       xlabel="Days",
       ylabel="Items sold")

# 5. Save and show (saves the whole figure)
fig.savefig("images/sample-plot.png")
```

# pyplot vs matplotlib OO method
* when plotting sth quickly better to use pyplot
* when plotting complex / advanced data use OO method

```python
over50 = heart_disease[heart_disease["age"]>50]
over50
```

```python
# pyplot method
over50.plot(kind="scatter",
            x='age',
            y='chol',
            c='target') #c is short for color
```

```python
# oo method
fig, ax = plt.subplots(figsize = (10,6))
over50.plot(kind = 'scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
ax.set_xlim([45,80])
```

```python
# OO method fromscratch
fig, ax = plt.subplots(figsize= (10.,6))

# Plot the data
scatter = ax.scatter(x=over50["age"],
                     y=over50["chol"],
                     c=over50["target"])
# Customize the plot
ax.set_title('Heart Disease and Cholestrol Levels')
ax.set_xlabel("Age")
ax.set_ylabel("Cholestrol");
ax.legend(*scatter.legend_elements(), title="Target")

# Add a horizontal line indicating mean cholestrol
ax.axhline(over50.chol.mean(), linestyle="--")

```

```python
# plot of age, chol, thalach
fig, (ax0,ax1) = plt.subplots(nrows=2,ncols=1,figsize=(10,10));

# add data to ax0
scatter = ax0.scatter(x=over50.age,
                     y=over50.chol,
                     c=over50.target);

# Customize ax0
ax0.set(title="Heart Disease and Cholestrol",
            xlabel= "Age",
            ylabel = "Cholestrol");
ax0.legend(*scatter.legend_elements(),title="Target")

# add horizontal line indicating mean cholestrol level
ax0.axhline(y=over50.chol.mean(),linestyle="--");


# Add data to ax1
scatter1 = ax1.scatter(x=over50.age,
                       y=over50.thalach,
                       c=over50.target);
# customize ax1
ax1.set(title="Heart Disease and Thalach",
        xlabel="Age",
        ylabel="Max Heart Rate");
ax1.legend(*scatter.legend_elements(),title="Target")

# Add horizontal line
ax1.axhline(y=over50.thalach.mean(),linestyle="--");

# add title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=28, fontweight='bold')
```

```python
# set style
plt.style.use("seaborn-v0_8-whitegrid")

# OO method fromscratch
fig, ax = plt.subplots(figsize= (10,6))

# Plot the data
scatter = ax.scatter(x=over50["age"],
                     y=over50["chol"],
                     c=over50["target"],
                     cmap='winter')
# Customize the plot
ax.set_title('Heart Disease and Cholestrol Levels')
ax.set_xlabel("Age")
ax.set_ylabel("Cholestrol");
ax.legend(*scatter.legend_elements(), title="Target")

# Add a horizontal line indicating mean cholestrol
ax.axhline(over50.chol.mean(), linestyle="--");
```

```python
# plot of age, chol, thalach
fig, (ax0,ax1) = plt.subplots(nrows=2,ncols=1,figsize=(10,10), sharex=True);

# add data to ax0
scatter = ax0.scatter(x=over50.age,
                     y=over50.chol,
                     c=over50.target,
                     cmap='winter');

# Customize ax0
ax0.set(title="Heart Disease and Cholestrol",
            xlabel= "Age",
            ylabel = "Cholestrol");

# Add a legend
ax0.legend(*scatter.legend_elements(),title="Target")

# add horizontal line indicating mean cholestrol level
ax0.axhline(y=over50.chol.mean(),linestyle="--");


# Add data to ax1
scatter1 = ax1.scatter(x=over50.age,
                       y=over50.thalach,
                       c=over50.target,
                       cmap='winter');
# customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate");

# add a legend
ax1.legend(*scatter.legend_elements(),title="Target")

# Add horizontal line
ax1.axhline(y=over50.thalach.mean(),linestyle="--");

# add title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=28, fontweight='bold');
```
